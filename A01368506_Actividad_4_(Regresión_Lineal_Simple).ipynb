```python
nmx_numero['host_response_rate'] = nmx_numero['host_response_rate'].str.rstrip('%').astype('float')
nmx_numero['host_acceptance_rate'] = nmx_numero['host_acceptance_rate'].str.rstrip('%').astype('float')
nmx_numero['price'] = nmx_numero['price'].str.replace('$', '').str.replace(',', '').astype('float')

nmx2_numero['host_response_rate'] = nmx2_numero['host_response_rate'].str.rstrip('%').astype('float')
nmx2_numero['host_acceptance_rate'] = nmx2_numero['host_acceptance_rate'].str.rstrip('%').astype('float')
nmx2_numero['price'] = nmx2_numero['price'].str.replace('$', '').str.replace(',', '').astype('float')
```

```python
y= nmx_numero
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
dnmex= nmx_numero[(y<=Limite_Superior)&(y>=Limite_Inferior)]
dnmex
```

```python
valores_nulos=dnmex.isnull().sum()
valores_nulos
```

```python
limpios=dnmex.copy()
limpios=limpios.fillna(round(dnmex.mean(),1))
limpios
```

```python
valores_nulos=limpios.isnull().sum()
valores_nulos
```

```python
# Aquí se realizó la correlación que existe en cada habitación

Columnas = ["host_acceptance_rate", "host_response_rate",
            "review_scores_location", "review_scores_cleanliness",
            "price", "availability_365", "number_of_reviews",
            "reviews_per_month", "review_scores_communication"]
Variables = [("host_acceptance_rate", "host_response_rate"),
                ("review_scores_location", "review_scores_cleanliness"),
                ("host_acceptance_rate", "price"),
                ("availability_365", "number_of_reviews"),
                ("host_acceptance_rate", "number_of_reviews"),
                ("reviews_per_month", "review_scores_communication")]

for variable1, variable2 in Variables:
    correlation = limpios[variable1].corr(limpios[variable2])
    print(f"Correlación entre {variable1} y {variable2}: {correlation}")
    plt.figure(figsize=(8, 6))
    plt.scatter(limpios[variable1], limpios[variable2], alpha=0.5)
    plt.title(f'Scatter Plot: {variable1} vs {variable2}')
    plt.xlabel(variable1)
    plt.ylabel(variable2)
    plt.show()
```
